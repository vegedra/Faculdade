```python
# Imports
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import r2_score, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
# Criacao do dataframe
df = pd.DataFrame({
    "horas_estudo": [1,2,3,10,11,12]  # Conjunto
})
print(df)
```

```python
# Definindo x
x = df[["horas_estudo"]]
```

```python
# Cria o kmeans - qtde de grupos, semente pra geracao de centroide, iteracoes
kmeans = KMeans(n_clusters = 2, random_state = 42, n_init = 10)
```

```python
# Aplicacao
kmeans.fit(x)
```

```python
# Exibicao
df["cluster"] = kmeans.labels_
print(df)
print("\nCentroides:\n", kmeans.cluster_centers_)  # Exibe os centroides
```

```python
# Atividade 1: criar dataframe com nota de 6 alunos em matematica e estatistica ao mesmo tempo

# Criacao do dataframe
df = pd.DataFrame({
    "notas_mat": [6,7,2,1,10,9],
    "notas_est": [7,2,3,1,10,8]
})

x = df[["notas_mat", "notas_est"]]

# Cria o kmeans - qtde de grupos, semente pra geracao de centroide, iteracoes
kmeans = KMeans(n_clusters = 2, random_state = 42, n_init = 10)

kmeans.fit(x)

df["cluster"] = kmeans.labels_
print(df)
```

```python
# Atividade 2:
# Criar dataset com 300 observacoes contendo:
# 3 variaveis numericas (a minha escolha)
# 1 variavel textual aleatoria
# 1 variavel resposta (nota_final)

# Faca uma regressao linear e invente algo para uma reg. logistica

# ==== REG. LINEAR =====
print("Reg Linear")
np.random.seed(42)
n = 300

# Criacao do dataframe
df = pd.DataFrame({
    "notas_mat": np.random.normal(70, 10, n),
    "notas_est": np.random.normal(65, 12, n),
    "notas_log": np.random.normal(75, 8, n),
    "periodo": np.random.choice(["Manha", "Tarde", "Noite"], n)  # Nao usa pois eh texto
})

df["nota_final"] = (
    0.3 * df["notas_mat"] +
    0.2 * df["notas_est"] +
    0.1 * df["notas_log"] +
    np.random.normal(0, 5, n)
)
print(df)

# Variavel resposta
x = df[["notas_mat", "notas_est", "notas_log"]]
y = df["nota_final"]

# Reg Linear
model = LinearRegression()
model.fit(x, y)

df["nota_prevista"] = model.predict(x)
print(df[["nota_final", "nota_prevista"]].head(10))

r2 = r2_score(y, df["nota_prevista"])
print("\nAcertou:\n", r2)

# ====== REG. LOGISTICA ======
print("Reg. logistica")
df["aprovado"] = (df["nota_final"] >= 70).astype(int)

x = df[["notas_mat", "notas_est", "notas_log"]]
y = df["aprovado"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

log_model = LogisticRegression(max_iter=1000)
log_model.fit(x_train, y_train)

y_pred = log_model.predict(x_test)
probs = log_model.predict_proba(x_test)

cm = confusion_matrix(y_test, y_pred)

print("\n", cm)
acc = accuracy_score(y_test, y_pred)
print("\nAcuracia:", acc)
```
